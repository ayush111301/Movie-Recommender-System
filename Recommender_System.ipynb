# **Movie Recommender System Working**




**Importing Required Libraries and Functions:**

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

**Loading of Data-set**

```python
movies_frame=pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
movies_frame.shape
```

```python
movies_frame.head()
```

```python
movies_frame
```

```python
movies_frame.head()
```

**Feature Selection:**

```python
relevant_features=['genres','keywords','popularity','title','runtime','tagline']
```

```python
for feature in relevant_features:
  movies_frame[feature]=movies_frame[feature].fillna('')
```

```python
high_end_features= movies_frame['genres']+ ' ' +movies_frame['keywords']+ ' ' +str(movies_frame['popularity'])+ ' ' +movies_frame['title']+ ' ' +str(movies_frame['runtime'])+ ' ' +movies_frame['tagline']
```

```python
print(high_end_features)
```

**Converting Text into Numbers Using TfidfVectorizer**

```python
number_vector=TfidfVectorizer()
```

```python
relevant_vectors=number_vector.fit_transform(high_end_features)
```

```python
print(relevant_vectors)
```

**Cosine Similarity Use:**

```python
similarity_score=cosine_similarity(relevant_vectors)
```

```python
print(similarity_score)
```

```python
print(similarity_score.shape)
```

**Getting User Input:**

```python
user_movie_input=input("Enter your most liked movie name: ")
```

```python
movies_title_list=movies_frame['title'].tolist()
```

```python
print(movies_title_list)
```

```python
closest_match=difflib.get_close_matches(user_movie_input, movies_title_list)
```

```python
print(closest_match)
```

**Making a Column Named index for the Dataframe:**

```python
for data in movies_frame.columns:
  print(data)
```

```python
movies_frame.insert(0,'index',range(0,len(movies_frame)))
```

```python
movies_frame.set_index("index")
```

```python
for data in movies_frame.columns:
  print(data)
```

**Getting Index of Movie Entered by User:**

```python
movie_index=movies_frame[movies_frame.title == closest_match[0]]['index'].values[0]
```

```python
print(movie_index)
```

**Making a List of all Similarity Scores:**

```python
similarity = list(enumerate(similarity_score[movie_index]))
print(similarity)
```

**Getting Movies which are similar to the movie user entered.**

```python
sort_movies_similarity=sorted(similarity, key= lambda x:x[1], reverse=True)
print(sort_movies_similarity)
```

**Printing Names of the Movies having high similarity score based on the user input.**

```python
print("Recommended Movies: ")
i=1
for movie in sort_movies_similarity:
  index=movie[0]
  index_title=movies_frame[movies_frame.index==index]['title'].values[0]
  if i<=15:
    print(i,'.',index_title)
    i=i+1
```

**Movie Recommender Sytem:**

```python
user_movie_input=input("Enter your most liked movie name: ")

movies_title_list=movies_frame['title'].tolist()

closest_match=difflib.get_close_matches(user_movie_input, movies_title_list)

movie_index=movies_frame[movies_frame.title == closest_match[0]]['index'].values[0]

similarity = list(enumerate(similarity_score[movie_index]))

sort_movies_similarity=sorted(similarity, key= lambda x:x[1], reverse=True)

print("Recommended Movies: ")
i=1
for movie in sort_movies_similarity:
  index=movie[0]
  index_title=movies_frame[movies_frame.index==index]['title'].values[0]
  if i<=20:
    print(i,'.',index_title)
    i=i+1
```
